# Analysis

```python
pj = pr.load('test_Pt_210')
```

```python
pj['output']
```

```python
plt.imshow(pj['output/total_FIM/21.5'])
```

```python
plt.plot(np.mean(pj['output/z_resolved_FIM/21.5'],axis=(0,1)))
plt.ylabel("FIM contribution")
```

```python
pot,_,_,cell = fim.potential(pj.input['simulator_dict']).potential_cell()
ex = fim.extra_waves(pj.input['simulator_dict'],pot=pot)
```

```python
xp =np.linspace(0,50,250)
yp= np.linspace(0,50,250)
#xp = np.linspace(0, fim_simulation.cell[0, 0], fim_simulation.Nx)
#yp = np.linspace(0, fim_simulation.cell[1, 1], fim_simulation.Ny)
FIM_image_ex_all=np.zeros([xp.shape[0],yp.shape[0]],dtype=np.complex128)

prho_rec_ex_all= np.fft.ifft2(pj['output/total_FIM/21.5'])

for ix in range(xp.shape[0]):
    for iy in range(yp.shape[0]):
        x=xp[ix]
        y=yp[iy]
        phase_1=np.exp(-1j*(ex.gk_1[:,0]*x+ex.gk_1[:,1]*y))
        phase_2= np.exp(-1j*(ex.gk_2[:,0]*x+ex.gk_2[:,1]*y))
        phase=np.outer(phase_1,phase_2)
        FIM_image_ex_all[ix,iy] =np.sum(prho_rec_ex_all.flatten()*phase.flatten())
```

```python
BOHR_TO_AA = 0.529177210903
```

```python
vmax_lev=np.max(FIM_image_ex_all.real)
xp_AA = xp*BOHR_TO_AA
yp_AA = yp*BOHR_TO_AA
plt.contourf(xp_AA,yp_AA,FIM_image_ex_all.real.T,vmin=0,vmax=vmax_lev,levels=np.linspace(0,vmax_lev,41))
plt.rcParams['font.size'] = '16'
plt.rcParams['font.family'] ='serif'
plt.xlabel('x ($\AA$)')
#plt.xticks([0,5,10,15,20,25])
plt.ylabel('y ($\AA$)')
plt.colorbar()
```

```python
pj.get_structure
struct = pj.structure
struct.plot3d()
```

```python
new_str = struct.repeat([int(np.ceil(xp_AA.max()/struct.positions[:,0].max())),int(np.ceil(yp_AA.max()/struct.positions[:,1].max())),1])
scatter_mask = new_str.positions[:, 2] > 11          
```

```python
vmax_lev=np.max(FIM_image_ex_all.real)
xp_AA = xp*BOHR_TO_AA
yp_AA = yp*BOHR_TO_AA
plt.contourf(xp_AA,yp_AA,FIM_image_ex_all.real.T,vmin=0,vmax=vmax_lev,levels=np.linspace(0,vmax_lev,41))
plt.rcParams['font.size'] = '16'
plt.rcParams['font.family'] ='serif'
plt.xlabel('x ($\AA$)')
#plt.xticks([0,5,10,15,20,25])
plt.ylabel('y ($\AA$)')
plt.colorbar()
plt.scatter(new_str.positions[scatter_mask, 0], new_str.positions[scatter_mask, 1],
                           c=new_str.get_atomic_numbers()[scatter_mask], cmap='rainbow', edgecolors='k', s=5*new_str.positions[scatter_mask,2])
plt.xlim([xp_AA.min(),xp_AA.max()])
plt.ylim([yp_AA.min(),yp_AA.max()])
plt.show();
```
